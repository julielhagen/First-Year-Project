```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
import os
from math import floor, ceil, nan
from scipy.ndimage import rotate, center_of_mass
from cut import cut_mask
from prep_image import prep_mask
from photutils.centroids import centroid_com
```

```python
def midpoint(mask):
    '''Find midpoint of mask array.'''
    row_mid = mask.shape[0] / 2
    col_mid = mask.shape[1] / 2
    return row_mid, col_mid

def get_centroid(mask):
    '''Find centroid of mask array.'''
    mask = np.where(mask == 0, nan, mask)
    return centroid_com(mask)

def plot_midpoint(mask):
    '''Cut mask and plot midpoint.'''
    cutted_mask = cut_mask(mask)
    y, x = midpoint(cutted_mask)
    plt.imshow(cutted_mask, cmap="gray")
    plt.axvline(x = x, color = "r")
    plt.axhline(y = y, color = "r")

def plot_centroid(mask):
    '''Cut mask and plot centroid.'''
    cutted_mask = cut_mask(mask)
    y, x = get_centroid(cutted_mask)
    plt.imshow(cutted_mask, cmap="gray")
    plt.axvline(x = x, color = "r")
    plt.axhline(y = y, color = "r")
```
