```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.color import rgb2gray
from skimage.filters import gaussian
from skimage.segmentation import active_contour
from skimage import data
```

```python
file_im = 'test_im.jpeg.png'
im = plt.imread(file_im)
im = rgb2gray(im)

plt.imshow(im, cmap='gray')
```

```python
#Circle
s = np.linspace(0, 2 * np.pi, 200)
r = 200 + 100 * np.sin(s)
c = 200 + 100 * np.cos(s)
init = np.array([r, c]).T
```

```python
#Smoothing
im2 = gaussian(im, 3, preserve_range=False)
plt.imshow(im2, cmap='gray')
```

```python
#Snake
snake = active_contour(im2, init, w_line=0)
```

```python
fig, axes = plt.subplots(figsize=(7,7))
axes.imshow(im2, cmap=plt.cm.gray)
axes.plot(init[:, 1], init[:, 0], '--r', lw=3)
axes.plot(snake[:, 1], snake[:, 0], '-b', lw=3)
plt.show()
```
